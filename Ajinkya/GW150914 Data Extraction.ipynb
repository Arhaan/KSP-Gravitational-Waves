```python
import numpy as np
import matplotlib.pyplot as plt
import gwpy 
from gwpy.timeseries import TimeSeries
from gwosc.datasets import event_gps
```

## Obtaining the GPS time of GW150914

```python
gps = event_gps('GW150914')
print("The event gps is {} s".format(gps))
```

## Fetching the timeseries for GW150914

```python
segment = (int(gps)-16,int(gps)+16)
strain_data = TimeSeries.fetch_open_data('L1',*segment,verbose=True)
print(strain_data[strain_data])
```

##  Time domain plot of GW150914 data

```python
strain_data.plot()
```

## Bandpassing and plotting the data 

```python
filtered_data = strain_data.bandpass(10,512)
filtered_data.plot()
```

## Saving the data to a text file 

```python
data = np.array(filtered_data)
np.savetxt('GW150914_data.txt',data)
```
